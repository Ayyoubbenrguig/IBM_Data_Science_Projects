# **Régression Linéaire**

### 1. **Régression Linéaire : Concept Théorique**
La régression linéaire est une méthode statistique qui cherche à modéliser la relation entre une variable dépendante \( y \) et une ou plusieurs variables indépendantes \( x \). Elle tente de prédire la valeur de \( y \) en fonction de \( x \) à l'aide d'une équation de droite (ou d'un plan hyperdimensionnel pour plusieurs variables) en minimisant l'erreur entre les prédictions et les valeurs réelles.

### 2. **Formule de la Régression Linéaire Simple**
En régression linéaire simple, nous n’avons qu’une seule variable indépendante \( x \). L’équation est :
\[
y = w_0 + w_1 x + \epsilon
\]
où :
- $( y )$ est la variable dépendante,
- $( w_0 )$ est l'ordonnée à l'origine (interception),
- $( w_1 )$ est le coefficient de pente (poids) associé à \( x \),
- $\epsilon $  est le terme d'erreur (ou résidu) qui représente l'écart entre la valeur observée et la valeur prédite.

Le but est de trouver les valeurs de $( w_0 $) et $( w_1 )$ qui minimisent l'erreur quadratique moyenne (Mean Squared Error, MSE).

### 3. **Formule de la Régression Linéaire Multiple**

En régression linéaire multiple, nous avons plusieurs variables indépendantes $x_1, x_2, \dots, x_n$. L’équation devient alors :

$$
y = w_0 + w_1 x_1 + w_2 x_2 + \dots + w_n x_n + \epsilon
$$

où chaque $w_i$ représente le poids (ou coefficient) pour chaque variable indépendante $x_i$.

La formule générale pour le modèle de régression multiple en notation matricielle est :

$$
\mathbf{y} = \mathbf{X} \mathbf{w} + \epsilon
$$

où :
- $\mathbf{y}$ est le vecteur des valeurs de sortie,
- $\mathbf{X}$ est la matrice des valeurs des variables indépendantes,
- $\mathbf{w}$ est le vecteur des poids,
- $\epsilon$ est le vecteur des termes d'erreur.

### 4. **Méthode des Moindres Carrés**

Pour estimer les valeurs optimales des paramètres $w$, on utilise la méthode des moindres carrés. L'objectif est de minimiser la fonction de coût suivante :

$$
J(w) = \frac{1}{N} \sum_{i=1}^{N} \left(y_i - (\mathbf{X}_i \cdot \mathbf{w})\right)^2
$$

où :
- $J(w)$ est l'erreur quadratique moyenne (MSE),
- $N$ est le nombre d'observations,
- $y_i$ est la valeur réelle de l'observation $i$,
- $\mathbf{X}_i \cdot \mathbf{w}$ est la prédiction pour l'observation $i$.

La solution des moindres carrés pour $w$ est :

$$
\mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

### 5. **Implémentation en Python**

#### Régression Linéaire Simple
Utilisons la bibliothèque `scikit-learn` pour implémenter une régression linéaire simple.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Données d'exemple

```

### 6. **Limitations de la Régression Linéaire**
- **Linéarité** : La régression linéaire suppose une relation linéaire entre les variables indépendantes et la variable dépendante. Elle n'est pas adaptée aux relations non linéaires sans transformation des données.
- **Sensibilité aux outliers** : Les valeurs extrêmes peuvent affecter fortement les prédictions.
- **Colinéarité** : En régression multiple, la colinéarité entre les variables indépendantes peut déstabiliser les coefficients et rendre les estimations imprécises.
  
La régression linéaire est un modèle de base en apprentissage supervisé qui fonctionne bien pour des relations linéaires simples.

$y = Xw + \epsilon$

Bien sûr ! Voici les formules de régression logistique réécrites avec des symboles LaTeX encadrés par `$` pour chaque expression mathématique :

### 1. Fonction Logistique (Sigmoïde)

La fonction sigmoïde, qui transforme l’entrée linéaire en probabilité, est donnée par :

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

où $z$ est une combinaison linéaire des variables indépendantes :

\[
z = w_0 + w_1 x_1 + w_2 x_2 + \dots + w_n x_n = \mathbf{X} \cdot \mathbf{w}
\]

### 2. Fonction de Probabilité

La probabilité que la sortie $y$ appartienne à la classe 1 (positive) est donnée par :

\[
P(y = 1 | X) = \sigma(\mathbf{X} \cdot \mathbf{w}) = \frac{1}{1 + e^{-(\mathbf{X} \cdot \mathbf{w})}}
\]

La probabilité pour la classe 0 est donc :

\[
P(y = 0 | X) = 1 - \sigma(\mathbf{X} \cdot \mathbf{w})
\]

### 3. Fonction de Perte : Log Loss (Erreur Logarithmique)

La fonction de perte utilisée pour la régression logistique est appelée **log loss** ou erreur logarithmique, formulée comme suit :

$
L(w) = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$

où $\hat{y}_i = \sigma(\mathbf{X}_i \cdot \mathbf{w})$ est la probabilité prédite pour l’observation $i$, et $N$ est le nombre total d'observations.

### 4. Estimation des Coefficients (Méthode des Moindres Carrés)

En régression logistique, les coefficients peuvent être calculés par la maximisation de la vraisemblance. La fonction de coût peut être minimisée pour obtenir les valeurs optimales des paramètres $\mathbf{w}$.

La **régression logistique** est un modèle d’apprentissage supervisé largement utilisé pour les problèmes de classification binaire (deux classes), bien qu’elle puisse être étendue à des classifications multiclasse. Contrairement à la régression linéaire, la régression logistique est conçue pour prédire des probabilités pour des classes distinctes en utilisant une fonction logistique (ou sigmoïde).

### 1. Théorie et Formules de la Régression Logistique

#### Fonction Logistique (Sigmoïde)

La régression logistique repose sur la fonction **sigmoïde** pour transformer une entrée linéaire en une probabilité. Cette fonction est définie par :

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

où \( z \) est une combinaison linéaire des variables indépendantes. En régression logistique, \( z \) est souvent écrit comme :

\[
z = w_0 + w_1 x_1 + w_2 x_2 + \dots + w_n x_n = \mathbf{X} \cdot \mathbf{w}
\]

où \( w_i \) représente les coefficients (ou poids) associés aux variables \( x_i \), et \( \mathbf{X} \) est le vecteur des valeurs des variables indépendantes.

#### Fonction de Probabilité

Le modèle de régression logistique prédit la probabilité que la sortie \( y \) appartienne à la classe 1 (ou classe positive) :

\[
P(y = 1 | X) = \sigma(\mathbf{X} \cdot \mathbf{w}) = \frac{1}{1 + e^{-(\mathbf{X} \cdot \mathbf{w})}}
\]

La probabilité que \( y \) appartienne à la classe 0 est donc :

\[
P(y = 0 | X) = 1 - \sigma(\mathbf{X} \cdot \mathbf{w})
\]

#### Fonction de Perte : Log Loss (Erreur Logarithmique)

Pour ajuster les poids \( w \), la régression logistique utilise la **fonction de perte logistique** (log loss), qui cherche à maximiser la vraisemblance des données observées :

\[
L(w) = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
\]

où \( \hat{y}_i = \sigma(\mathbf{X}_i \cdot \mathbf{w}) \) est la probabilité prédite pour l’observation \( i \), et \( N \) est le nombre total d'observations.

### 2. Interprétation Statistique

La régression logistique peut être interprétée comme un **modèle de probabilité**. Les coefficients \( w \) indiquent l'effet des variables indépendantes \( X \) sur la logit (logarithme du rapport de chances) de la variable dépendante \( y \). Autrement dit, chaque coefficient \( w_i \) montre comment un changement unitaire dans \( x_i \) affecte la probabilité d’appartenir à la classe cible.


### Limitations

- **Non-linéarité des données :** La régression logistique ne fonctionne pas bien pour les relations non linéaires sans transformation préalable des variables.
- **Outliers :** Les valeurs aberrantes peuvent affecter la stabilité du modèle.
- **Multicolinéarité :** En présence de variables fortement corrélées, les coefficients peuvent être instables.

### Applications de la Régression Logistique

- **Détection de spam** : Classifier les emails comme spam ou non-spam.
- **Diagnostic médical** : Prédire si un patient a une certaine maladie (oui/non).
- **Prévision de comportements clients** : Prédire si un utilisateur achètera un produit ou non.

En résumé, la régression logistique est un modèle de base mais efficace pour des problèmes de classification binaire. Elle est simple à interpréter et peut être un bon point de départ pour de nombreuses applications.

### 3. Implémentation en Python avec un Exemple

Supposons que nous ayons un ensemble de données fictif contenant deux variables : **âge** et **salaire**, et que nous voulons prédire si un individu achètera un produit (1) ou non (0).



````python
#### Importation des Bibliothèques et Création du Dataset

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Création d'un dataset fictif
data = pd.DataFrame({
    'age': [22, 25, 47, 52, 46, 56, 55, 60, 62, 61],
    'salary': [25000, 32000, 52000, 110000, 84000, 150000, 95000, 200000, 210000, 150000],
    'bought': [0, 0, 1, 1, 1, 1, 0, 1, 1, 1]
})
````

````python
# Augmentation de cette data

# Générer des valeurs aléatoires en fonction des statistiques des données d'origine

np.random.seed(0)  # Assure la reproductibilité

# Âge : Moyenne et écart-type basés sur les valeurs d'origine
age_mean = data['age'].mean()
age_std = data['age'].std()
ages = np.random.normal(age_mean, age_std, 1000).astype(int)

# Salaire : Moyenne et écart-type basés sur les valeurs d'origine
salary_mean = data['salary'].mean()
salary_std = data['salary'].std()
salaries = np.random.normal(salary_mean, salary_std, 1000).astype(int)

# Achats : Répartition proportionnelle des 0 et 1
boughts = np.random.choice([0, 1], 1000, p=[0.3, 0.7])

# Création du DataFrame avec 100 observations
augmented_data = pd.DataFrame({
    'age': ages,
    'salary': salaries,
    'bought': boughts
})
````

````python

# Affichage des premières lignes du DataFrame
print(augmented_data.head())
````

````python
# Variables indépendantes (features) et dépendante (target)
X = data[['age', 'salary']]
y = data['bought']

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

````python
#### Entraînement du Modèle de Régression Logistique
# Initialisation et entraînement du modèle
model = LogisticRegression()
model.fit(X_train, y_train)
````

````python

#Prédictions et Évaluation du Modèle
# Prédictions
y_pred = model.predict(X_test)

# Calcul de la précision et affichage des résultats
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
````

````python
# Affichage des coefficients

coefficients = model.coef_
intercept = model.intercept_

print("Intercept:", intercept)
print("Coefficients:", coefficients)
````

Chaque coefficient montre l'effet d'une unité d'augmentation de la variable correspondante sur la logit (logarithme du rapport de chances) de l’achat du produit.

# **Modèle KNN**

Le **K-Nearest Neighbors (KNN)**, ou méthode des K plus proches voisins, est un modèle d’apprentissage supervisé utilisé principalement pour des tâches de classification et, dans certains cas, de régression. Il est apprécié pour sa simplicité et sa capacité à bien fonctionner avec des données de taille modérée. Voici une vue d’ensemble de son principe, fonctionnement, théorie, et un exemple de code avec cas d’étude.

### Principe du KNN

Le KNN repose sur l’idée que des observations similaires (ou "voisines") sont susceptibles d’appartenir à la même classe. Le modèle utilise la distance entre les points de données pour prédire la classe ou la valeur cible d’une nouvelle observation en fonction des "K" observations les plus proches.

### Fonctionnement du KNN

1. **Choix de K** : On fixe un nombre \( K \), qui est le nombre de voisins à considérer.
2. **Calcul des distances** : Pour chaque point à prédire, on calcule sa distance avec tous les autres points du dataset (le plus souvent, la distance euclidienne).
3. **Sélection des K voisins les plus proches** : On sélectionne les \( K \) points les plus proches de l’observation cible.
4. **Vote majoritaire (ou moyenne)** :
   - **Classification** : Pour une classification, le modèle attribue la classe la plus fréquente parmi les K voisins.
   - **Régression** : Pour une régression, le modèle calcule la moyenne des valeurs cibles des K voisins.

### Théorie : Distance Euclidienne

La distance euclidienne est la mesure de similarité la plus courante pour le KNN. La distance entre deux points $A = x_1, y_1$ et $ B = (x_2, y_2)$est :


$d(A, B) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$


Dans le cas de données de dimension \( n \), la distance entre deux points $\ X = (x_1, x_2, \dots, x_n) \ et \ Y = (y_1, y_2, \dots, y_n) \$ est :


$d(X, Y) = \sqrt{\sum_{i=1}^{n} (y_i - x_i)^2}$



### Code pour le KNN en Python

Voici un exemple de code implémentant le KNN pour un problème de classification en utilisant Scikit-Learn.

#### Cas d’étude : Classification d'Iris

L'objectif est de classer des fleurs en trois espèces (setosa, versicolor, et virginica) en fonction de leurs mesures.


### Limites et défis du KNN

- **Lenteur avec des datasets volumineux** : Comme il doit calculer les distances entre le point de test et tous les points d'entraînement, le KNN peut devenir lent pour de grandes bases de données.
- **Sensibilité au choix de K** : Un petit \( K \) peut rendre le modèle sensible au bruit, tandis qu'un \( K \) trop grand peut le rendre trop généralisé.
- **Influence des caractéristiques** : Les caractéristiques avec des valeurs de grande amplitude peuvent dominer les autres. Normaliser les données est essentiel pour de bonnes performances.

### Résumé

Le KNN est une méthode intuitive et facile à implémenter qui repose sur le calcul de distances entre les observations. Bien qu’efficace pour des datasets modérés et lorsque les classes sont bien séparées, il devient moins performant pour les datasets plus complexes ou de grande taille.

Cet exemple montre comment utiliser le KNN pour une tâche de classification. Pour la régression, on utilise `KNeighborsRegressor` de Scikit-Learn de manière similaire.

````python
# Import des bibliothèques
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# Chargement du dataset d'Iris
iris = load_iris()
X = iris.data  # Caractéristiques (longueur et largeur des sépales et pétales)
y = iris.target  # Espèce de la fleur
````

````python
iris.data
````

````python
iris.target
````

````python
# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
# Initialisation du modèle KNN avec K = 3
k = 20
knn = KNeighborsClassifier(n_neighbors=k)

# Entraînement du modèle
knn.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = knn.predict(X_test)

# Évaluation du modèle
accuracy = accuracy_score(y_test, y_pred)

print(f"Précision du modèle KNN avec K={k}: {accuracy:.2f}")
````

````python
# Tester différents K de 1 à 20
k_values = range(1, 21)
mean_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')  # Utilisation de la précision pour évaluer
    mean_scores.append(scores.mean())

# Affichage des résultats
import matplotlib.pyplot as plt
plt.plot(k_values, mean_scores, marker='o')
plt.xlabel('Valeur de K')
plt.ylabel('Précision moyenne')
plt.title('Choisir la meilleure valeur de K pour KNN')
plt.show()

# Identifier le meilleur K
best_k = k_values[mean_scores.index(max(mean_scores))]
print(f"Meilleure valeur de K : {best_k}")
````

# **Support vector machine**

### Support Vector Machine (SVM) : Principe, Fonctionnement, Théorie et Code

#### 1. **Principe de SVM :**
Le modèle Support Vector Machine (SVM) est un algorithme d'apprentissage supervisé utilisé principalement pour la classification, mais aussi pour la régression. Le principe fondamental d'un SVM est de trouver l'hyperplan qui sépare les différentes classes avec la marge maximale. Ce hyperplan est choisi de manière à maximiser la distance entre les points les plus proches de chaque classe, appelés **vecteurs de support**.

#### 2. **Théorie :**
L'objectif d'un SVM est de résoudre le problème suivant :

$\text{Maximiser } \frac{2}{\|\mathbf{w}\|}$

sous la contrainte :


Le modèle SVM cherche à maximiser la fonction suivante :

sous la contrainte :


$y_i (\mathbf{w} \cdot \mathbf{x_i} + b) \geq 1 \quad \text{pour chaque } i$


où :

- $\mathbf{x_i}$ est un vecteur de caractéristiques,
- $y_i$ est la classe correspondante de $\mathbf{x_i}$ (valeurs +1 ou -1),
- $\mathbf{w}$ est le vecteur normal à l'hyperplan,
- $b$ est le biais.

Le modèle SVM trouve donc l'hyperplan qui maximisera la distance entre les classes tout en respectant les contraintes ci-dessus.

#### 3. **Optimisation du SVM :**
La formulation classique du SVM peut être écrite sous forme d'un problème d'optimisation quadratique. En réarrangeant les équations, la fonction de coût à minimiser devient :

$\text{Minimiser } \frac{1}{2} \|\mathbf{w}\|^2$

avec des contraintes de classification pour chaque point de données, comme indiqué ci-dessus.

#### 4. **SVM avec noyau (Kernel Trick) :**
Un problème majeur avec SVM est que, dans les cas où les classes ne sont pas linéairement séparables, il devient difficile de trouver un hyperplan de séparation. Cependant, l'idée est d'appliquer un **truc de noyau (kernel trick)** qui permet de transformer les données dans un espace de caractéristiques de plus grande dimension, rendant ainsi les données linéairement séparables.

Les noyaux populaires incluent :
- **Noyau linéaire** : $ K(x, x') = x^T x'$
- **Noyau polynomial** : $ K(x, x') = (x^T x' + c)^d $
- **Noyau gaussien (RBF)** : $K(x, x') = \exp\left(-\frac{\|x - x'\|^2}{2\sigma^2}\right)$

#### 6. **Conclusion et Limites du SVM :**
- **Avantages** :
  - Efficace dans les espaces de grande dimension.
  - Peut être utilisé pour des problèmes linéaires et non linéaires grâce au noyau.
  - SVM garantit une solution optimale en maximisant la marge de séparation.

- **Limites** :
  - Moins efficace pour de très grands ensembles de données.
  - Le choix du noyau et des hyperparamètres (comme `C` et `gamma`) peut être délicat.
  - L'entraînement peut être coûteux en temps pour les grandes quantités de données.

SVM reste l'un des algorithmes les plus puissants pour la classification, en particulier pour les problèmes où la séparation des classes est nette et en grande dimension.

````python
##### b. **Code Python pour SVM :**

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Charger un jeu de données exemple (Iris)
data = datasets.load_iris()
X = data.data[:, :2]  # Utilisation des 2 premières features pour une visualisation facile
y = data.target

# On se limite à un problème binaire pour simplifier
X = X[y != 2]
y = y[y != 2]

# Séparation en données d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

````

````python
# Créer un modèle SVM avec noyau RBF (Radial Basis Function)
model = SVC(kernel='rbf', C=1, gamma=0.5)

# Entraîner le modèle
model.fit(X_train, y_train)


````

````python
# Prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Afficher les résultats
print("Classification Report :\n", classification_report(y_test, y_pred))
````

````python
# Visualisation des frontières de décision
xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Tracer la frontière de décision
plt.contourf(xx, yy, Z, alpha=0.75, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', cmap=plt.cm.coolwarm)
plt.title("Frontière de décision SVM")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
````

# **Arbre de décision**

Un arbre de décision est un algorithme de machine learning utilisé pour des tâches de classification et de régression. Il est interprétable et se construit en divisant récursivement les données en sous-groupes basés sur des caractéristiques, pour arriver à une prédiction ou une classification.

### 1. **Théorie**

Un arbre de décision suit une structure de "nœud racine", de "nœuds internes", et de "feuilles":
- **Nœud racine** : le point de départ de l'arbre, représentant la première décision.
- **Nœuds internes** : points de décision basés sur les caractéristiques.
- **Feuilles** : résultats finaux ou décisions (classes dans le cas de classification).

#### Critères de Division
L’arbre de décision se construit en choisissant la meilleure caractéristique à chaque niveau de l’arbre, selon certains critères comme :
1. **Entropie et Gain d’information** : L’entropie mesure l'incertitude dans les données. Le gain d’information évalue la réduction d'entropie après une division.
   
   $\text{Gain d'information} = \text{Entropie avant division} - \text{Entropie après division}$
   
   
2. **Indice de Gini** : utilisé dans l’algorithme CART (Classification and Regression Trees), mesure l'impureté d’un nœud.
   
   $\text{Indice de Gini} = 1 - \sum_{i=1}^{C} p_i^2$
   
   où $ p_i $ est la probabilité de classe $ i .$

### 2. **Construction de l'Arbre**

La construction d'un arbre de décision implique de choisir la meilleure caractéristique pour chaque division, selon un critère donné, jusqu'à ce que :
- Tous les échantillons appartiennent à la même classe (arbre pur),
- Une profondeur maximale est atteinte,
- Un seuil minimal de gain est atteint.

Les algorithmes courants pour construire des arbres de décision incluent **ID3**, **C4.5**, et **CART**.

### 3. **Limites**
- Les arbres de décision peuvent facilement sur-apprendre les données (overfitting), surtout si l'arbre est très profond.
- Ils sont sensibles aux variations dans les données.
- Ils peuvent être instables : de petits changements dans les données peuvent entraîner une structure d'arbre complètement différente.

### 5. **Exemple Concret (Mise en Situation)**

**Scénario** : Imaginons que vous travaillez pour une banque et que vous devez prédire si un client acceptera ou non une offre de prêt en fonction de son âge et de son salaire.

- **Âge** et **Salaire** seraient les caractéristiques de votre modèle, tandis que **Acceptance du prêt (0 ou 1)** serait l’étiquette.
- Vous créez un arbre de décision pour diviser les clients en fonction de l'âge et du salaire pour estimer la probabilité qu'ils acceptent le prêt.
- En observant l'arbre de décision obtenu, vous pouvez visualiser les règles. Par exemple, l’arbre pourrait indiquer que si l'âge est inférieur à 40 ans et le salaire supérieur à 50 000, alors il y a de fortes chances d'acceptation.

### Résumé

L’arbre de décision est simple, facile à interpréter et efficace pour la classification et la régression. Toutefois, il présente des limitations en termes de performance pour des jeux de données complexes, ce qui nécessite souvent l'utilisation de modèles avancés comme les forêts aléatoires et le gradient boosting pour améliorer la précision.

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Jeu de données exemple
data = pd.DataFrame({
    'age': [22, 25, 47, 52, 46, 56, 55, 60, 62, 61],
    'salary': [25000, 32000, 52000, 110000, 84000, 150000, 95000, 200000, 210000, 150000],
    'bought': [0, 0, 1, 1, 1, 1, 0, 1, 1, 1]
})

# Caractéristiques (X) et étiquette (y)
X = data[['age', 'salary']]
y = data['bought']

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
# Créer le modèle et l'entraîner
model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Faire des prédictions
y_pred = model.predict(X_test)

# Calculer la précision
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision de l'arbre de décision : {accuracy:.2f}")

# Visualisation de l'arbre (optionnel)
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=['age', 'salary'], class_names=['Non', 'Oui'], filled=True)
plt.show()
````

# **Random Forest**

La **Random Forest** (forêt aléatoire) est un algorithme d'apprentissage supervisé qui combine plusieurs arbres de décision pour améliorer la précision, la stabilité, et la capacité de généralisation. Cet algorithme est largement utilisé pour des tâches de classification et de régression en raison de sa résistance au surapprentissage et de sa capacité à bien s'adapter aux données complexes.

### 1. **Principe de la Random Forest**
La Random Forest repose sur un concept appelé **bagging** (ou **bootstrap aggregating**), qui consiste à :
1. Générer plusieurs échantillons aléatoires du jeu de données d'origine.
2. Créer un arbre de décision sur chaque échantillon.
3. Moyenne (pour la régression) ou vote majoritaire (pour la classification) des prédictions des arbres pour obtenir le résultat final.

Chaque arbre de décision est formé sur un sous-ensemble des données d'entraînement et utilise un sous-ensemble aléatoire des caractéristiques à chaque nœud, ce qui rend chaque arbre unique. En combinant les prédictions de plusieurs arbres indépendants, la Random Forest réduit la variance de la prédiction.

### 2. **Fonctionnement**

Le processus se divise en trois étapes principales :
1. **Échantillonnage aléatoire** : Chaque arbre est entraîné sur un échantillon aléatoire de l'ensemble de données d'origine, avec remplacement (bootstrap sampling).
2. **Sélection aléatoire des caractéristiques** : Lors de la construction de chaque arbre, un sous-ensemble aléatoire de caractéristiques est sélectionné pour chaque division, ce qui réduit la corrélation entre les arbres.
3. **Agrégation des prédictions** : Les prédictions des arbres sont combinées par :
   - **Vote majoritaire** pour la classification.
   - **Moyenne** pour la régression.

### 3. **Avantages et Inconvénients**

**Avantages** :
- **Robustesse au surapprentissage** : En combinant plusieurs arbres, la Random Forest réduit la variance et empêche le surajustement des données.
- **Performance élevée** : Bonne précision sur des jeux de données complexes.
- **Résistant aux valeurs aberrantes** : Comme chaque arbre utilise un sous-ensemble des données, les valeurs extrêmes ont moins d'impact.
- **Interprétabilité partielle** : Possibilité d'analyser l'importance des caractéristiques.

**Inconvénients** :
- **Complexité** : La Random Forest est plus lente et consomme plus de mémoire que les arbres de décision individuels.
- **Moins interprétable** : Bien qu'il soit possible d'observer l'importance des caractéristiques, la Random Forest est moins facile à interpréter qu'un seul arbre de décision.

### 4. **Formules Théoriques**

Pour un ensemble de $M$ arbres $T_1$, $T_2$, $\dots$, $T_M$ :
- **Classification** : La prédiction finale est déterminée par le vote majoritaire parmi les arbres :
  
  $\hat{y} = \text{mode} \{ T_1(x), T_2(x), \dots, T_M(x)$
  
- **Régression** : La prédiction finale est la moyenne des prédictions des arbres :
  
  $\hat{y} = \frac{1}{M} \sum_{m=1}^{M} T_m(x)$

### 6. **Cas d'Utilisation Concret : Prédiction d'Admission en Université**

**Contexte** : Vous travaillez avec des données d'étudiants et souhaitez prédire leur admission en fonction de leurs scores de test et de leurs GPA. Les caractéristiques peuvent inclure les scores des tests, GPA, activités parascolaires, etc.

- En utilisant une Random Forest, vous pouvez entraîner un modèle pour prédire l'admission.
- Les arbres individuels seront construits avec différents sous-ensembles de caractéristiques, ce qui permet de capturer différents aspects des données (certains arbres pourraient se concentrer sur le GPA, d'autres sur les scores de test).
- En combinant les prédictions des arbres, la Random Forest fournira une estimation robuste de la probabilité d'admission.

La Random Forest est particulièrement efficace pour des jeux de données de grande dimension ou lorsqu'il existe des corrélations non linéaires entre les caractéristiques. Elle est donc bien adaptée aux tâches de classification et de régression sur des données complexes.

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Jeu de données d'exemple
data = pd.DataFrame({
    'age': [22, 25, 47, 52, 46, 56, 55, 60, 62, 61],
    'salary': [25000, 32000, 52000, 110000, 84000, 150000, 95000, 200000, 210000, 150000],
    'bought': [0, 0, 1, 1, 1, 1, 0, 1, 1, 1]
})

# Caractéristiques et étiquettes
X = data[['age', 'salary']]
y = data['bought']

# Division des données
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Création et entraînement de la Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prédictions et évaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision de la Random Forest : {accuracy:.2f}")

# Importance des caractéristiques
importances = model.feature_importances_
for i, col in enumerate(X.columns):
    print(f"Importance de {col} : {importances[i]:.2f}")
````
